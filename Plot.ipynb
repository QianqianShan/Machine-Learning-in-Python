### Making Plots Using Matplotlib

```python
import numpy as np
import matplotlib
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import math
import matplotlib.pyplot as plt
```

```python
# Pdf of normal distribution
mu = 0 
sigma = 1
x = np.linspace(mu - 3 * sigma, mu + 3 * sigma, 50)
y = np.exp(-(x - mu) ** 2/(2 * sigma **2))/(math.sqrt(2 * math.pi) * sigma)
print(x.shape)
print(y.shape)
plt.plot(x, y, 'r-', x, y, 'go', linewidth=2, markersize=8) # lines and markers
# plt.plot(x, y, 'go', linewidth=2, markersize=8) 

plt.grid(True)
plt.show()
# help(plt.plot)
matplotlib.rcParams['font.sans-serif'] = [u'SimHei']  
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
# Loss function
x = np.array(np.linspace(start=-2, stop=3, num=1001, dtype=np.float))
y_logit = np.log(1 + np.exp(-x)) / math.log(2)  # logit 
y_boost = np.exp(-x)
y_01 = x < 0
y_hinge = 1.0 - x
y_hinge[y_hinge < 0] = 0

plt.plot(x, y_logit, 'r-', label='Logistic Loss', linewidth=2)
plt.plot(x, y_01, 'g-', label='0/1 Loss', linewidth=2)
plt.plot(x, y_hinge, 'b-', label='Hinge Loss', linewidth=2)
plt.plot(x, y_boost, 'm--', label='Adaboost Loss', linewidth=2)
plt.grid()
plt.legend(loc='upper right') # best, center ... 
plt.title("Loss Functions")
# plt.savefig('1.png')
plt.show()
```

```python
# x^x vs x 

# x ** x        x > 0
# (-x) ** (-x)  x < 0
def f(x):
    y = np.ones_like(x) # Initialization
    i = x > 0
    y[i] = np.power(x[i], x[i])
    i = x < 0
    y[i] = np.power(-x[i], -x[i])
    return y


x = np.linspace(-1.3, 1.3, 101)
y = f(x)
plt.plot(x, y, 'g-', label='x^x', linewidth=2)
plt.grid()
plt.legend(loc='upper center')
plt.show()
help(np.ones_like)
```

```python
x = np.arange(1, 0, -0.001)
y = (-3 * x * np.log(x) + np.exp(-(40 * (x - 1 / np.e)) ** 4) / 25) / 2
plt.figure(figsize=(5,7))
plt.plot(y, x, 'r-', linewidth=2)
plt.grid(True)
plt.show()
```

```python
# Heart 
t = np.linspace(0, 7, 100)
x = 16 * np.sin(t) ** 3
y = 13 * np.cos(t) - 5 * np.cos(2*t) - 2 * np.cos(3*t) - np.cos(4*t)
plt.plot(x, y, 'r-', linewidth=2)
plt.grid(True)
plt.show()
```

```python
t = np.linspace(0, 50, num=1000)
x = t*np.sin(t) + np.cos(t)
y = np.sin(t) - t*np.cos(t)
plt.plot(x, y, 'r-', linewidth=2)
plt.grid()
plt.show()
```

```python
 # Bar
# matplotlib.rcParams['font.sans-serif'] = [u'SimHei']  
# matplotlib.rcParams['axes.unicode_minus'] = False
x = np.arange(0, 10, 0.1)
y = np.sin(x)
plt.bar(x, y, width=0.04, linewidth=0.2)
plt.plot(x, y, 'r--', linewidth=2)
plt.title('Sin Curve')
plt.xticks(rotation=-60)
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.show()
```

```python
# PDF 
# Uniform 
x = np.random.rand(10000)
t = np.arange(len(x))
plt.hist(x, 30, color='b', alpha=0.5, label = "Uniform Distribution", rwidth = 0.8)
# plt.plot(t, x, 'r-', label= "Uniform Distribution")
plt.legend(loc='upper left')
plt.grid()
plt.show()
# help(plt.hist)
```

```python
# Central Limit Theorem 
t = 10000
a = np.zeros(10000) # 1000 zeros 
for i in range(t):
    a += np.random.uniform(-10, 10, 10000)
    a /= t

plt.hist(a, bins=30, color='g', alpha=0.5, normed=True)
plt.grid()
plt.show() # Most a values are around 0 

# np.random.uniform(-5, 5, 1000)
```

```python
# Poisson Distribution
x = np.random.poisson(lam = 5, size = 10000)
print(x)
pillar = 15
a = plt.hist(x, bins = pillar, normed = True, range = [0, pillar], color = 'b', alpha = 0.5)
plt.grid()
plt.show()
print(a)
print(a[0].sum()) # sum of pmf 
```

```python
# Histogram
from scipy.stats import norm, poisson

mu = 2
sigma = 3
data = mu + sigma * np.random.randn(1000)
h = plt.hist(data, bins = 30, normed = True, color='#a0a0ff')
x = h[1]
# print(x)
y = norm.pdf(x, loc = mu, scale = sigma)
plt.plot(x, y, 'r--', x, y, 'ro', linewidth=2, markersize=4)
plt.grid()
plt.show()
```

```python
# Interpolation 
from scipy.interpolate import BarycentricInterpolator
from scipy.interpolate import CubicSpline
import scipy

rv = poisson(5)
x1 = a[1]
y1 = rv.pmf(x1)
itp = BarycentricInterpolator(x1, y1)  
x2 = np.linspace(x.min(), x.max(), 50)
y2 = itp(x2)
cs = scipy.interpolate.CubicSpline(x1, y1)       
plt.plot(x2, cs(x2), 'm--', linewidth=5, label='CubicSpine')           
plt.plot(x2, y2, 'g-', linewidth=3, label='BarycentricInterpolator')   
plt.plot(x1, y1, 'r-', linewidth=1, label='Actural Value')             
plt.legend(loc='lower right')
plt.grid()
plt.show()
```
